```python
import pandas as pd
```

```python
notasIntermedias = pd.read_excel('intermediate_grades.xlsx',sep=',', sheet_name='Intermediate grades')

notasFinal1 = pd.read_excel('final_grades.xlsx',sep=',', sheet_name='Exam (First time)')
notasFinal2 = pd.read_excel('final_grades.xlsx',sep=',', sheet_name='Exam (Second time)')
```

```python
notasFinalPrimera=notasFinal1[['Student ID','TOTAL\n(100 points)']]

notasFinalPrimera.rename(columns={'Student ID': 'ID', 'TOTAL\n(100 points)': 'Nota primera oportunidad'}, inplace=True)
```

```python
notasFinalSegunda=notasFinal2[['Student ID','TOTAL\n(100 points)']]

notasFinalSegunda.rename(columns={'Student ID': 'ID', 'TOTAL\n(100 points)': 'Nota segunda oportunidad'}, inplace=True)

notasIntermedias.rename(columns={'Student Id': 'ID'}, inplace=True)
```

```python
objs = [notasFinalPrimera,notasFinalSegunda] 
notas = pd.concat(objs,ignore_index=True)
```

```python
lista = []
for i in range(1,116):
    lista.append(i)

data = {'ID': lista,'valor':lista}
df = pd.DataFrame(data,columns=['ID','valor'])
obj = [df,notasFinalPrimera,notasFinalSegunda]

res = pd.merge(df, notasFinalPrimera, on='ID', how='outer')
final = pd.merge(res, notasFinalSegunda, on='ID', how='outer')
notasCompleto = pd.merge(final, notasIntermedias, on='ID', how='outer')

notasExamenFinal = (notasCompleto[['ID','Session 2','Session 3','Session 4','Session 5','Session 6','Nota primera oportunidad','Nota segunda oportunidad']]).fillna(0)
```

```python
notasExamenFinal.to_csv('notasCurso.csv', index=False, header=True,encoding='utf-8')
```
