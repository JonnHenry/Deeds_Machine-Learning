```python
import pandas as pd
import os
import numpy as np
import statistics
```

```python
#Se lee las cabeceras necesarios para los archivos de las actividades de las sesiones
data = pd.read_csv('cabeceras.txt', sep=' ', header=None)
cabeceras = data[1].values.tolist()
print(cabeceras)
```

```python
#Se lee el archivo de los logs para poder obtener la información de las actividades si el estudiante ha estado en la session
sesiones = pd.read_csv('logs.txt', sep='\t')
sesiones
```

```python
archivo = pd.read_csv('Session 1/1', header=None, names=cabeceras)
pruebaArchivo = pd.read_csv('Session 1/4',header=None,names=cabeceras)
valor = len(archivo)+len(pruebaArchivo)
archivo.append(pruebaArchivo)
#pruebaArchivo
archivo
```

```python
existeArchivoSesion1=False
existeArchivoSesion2=False
existeArchivoSesion3=False
existeArchivoSesion4=False
existeArchivoSesion5=False
existeArchivoSesion6=False
columnas = ['session', 'student_Id', 'activity', 'idle_time']
recolectaDatos = {}
for index, row in sesiones.iterrows(): 
    
    if (row['Session 1']==1):
        pathArchivo = 'Session 1/'+str(row['Student Id'])
        if (os.path.isfile(pathArchivo)):
            session = pd.read_csv(pathArchivo, header=None, names=cabeceras)
            estudiante = session[columnas]
            
            for indice, fila in estudiante.iterrows():
                clave = (fila['activity']).split("_")[0].strip()
                idEstudiante = fila['student_Id']
                if (fila[idEstudiante].strip() in recolectaDatos[idEstudiante]):
                    recolectaDatos[idEstudiante][clave].append(fila['idle_time']/1000)
                else:
                    recolectaDatos[idEstudiante][clave] = []
                    recolectaDatos[idEstudiante][clave].append(fila['idle_time']/1000)

    if (row['Session 2']==1):
        pathArchivo = 'Session 2/'+str(row['Student Id'])
        if (os.path.isfile(pathArchivo)):
            session = pd.read_csv(pathArchivo, header=None, names=cabeceras)
            estudiante = session[columnas]
            
            for indice, fila in estudiante.iterrows():
                clave = (fila['activity']).split("_")[0].strip()
                idEstudiante = fila['student_Id']
                if (fila[idEstudiante].strip() in recolectaDatos[idEstudiante]):
                    recolectaDatos[idEstudiante][clave].append(fila['idle_time']/1000)
                else:
                    recolectaDatos[idEstudiante][clave] = []
                    recolectaDatos[idEstudiante][clave].append(fila['idle_time']/1000)
      

    if (row['Session 3']==1):
        pathArchivo = 'Session 3/'+str(row['Student Id'])
        if (os.path.isfile(pathArchivo)):
            session = pd.read_csv(pathArchivo, header=None, names=cabeceras)
            estudiante = session[columnas]
            
            for indice, fila in estudiante.iterrows():
                clave = (fila['activity']).split("_")[0].strip()
                idEstudiante = fila['student_Id']
                if (fila[idEstudiante].strip() in recolectaDatos[idEstudiante]):
                    recolectaDatos[idEstudiante][clave].append(fila['idle_time']/1000)
                else:
                    recolectaDatos[idEstudiante][clave] = []
                    recolectaDatos[idEstudiante][clave].append(fila['idle_time']/1000)
                    
    if (row['Session 4']==1):
        pathArchivo = 'Session 4/'+str(row['Student Id'])
        if (os.path.isfile(pathArchivo)):
            session = pd.read_csv(pathArchivo, header=None, names=cabeceras)
            estudiante = session[columnas]
            
            for indice, fila in estudiante.iterrows():
                clave = (fila['activity']).split("_")[0].strip()
                idEstudiante = fila['student_Id']
                if (fila[idEstudiante].strip() in recolectaDatos[idEstudiante]):
                    recolectaDatos[idEstudiante][clave].append(fila['idle_time']/1000)
                else:
                    recolectaDatos[idEstudiante][clave] = []
                    recolectaDatos[idEstudiante][clave].append(fila['idle_time']/1000)
                    
                    
    if (row['Session 5']==1):
        pathArchivo = 'Session 5/'+str(row['Student Id'])
        if (os.path.isfile(pathArchivo)):
            session = pd.read_csv(pathArchivo, header=None, names=cabeceras)
            estudiante = session[columnas]
            
            for indice, fila in estudiante.iterrows():
                clave = (fila['activity']).split("_")[0].strip()
                idEstudiante = fila['student_Id']
                if (fila[idEstudiante].strip() in recolectaDatos[idEstudiante]):
                    recolectaDatos[idEstudiante][clave].append(fila['idle_time']/1000)
                else:
                    recolectaDatos[idEstudiante][clave] = []
                    recolectaDatos[idEstudiante][clave].append(fila['idle_time']/1000)
                    
                    
                    
    if (row['Session 6']==1):
        pathArchivo = 'Session 6/'+str(row['Student Id'])
        if (os.path.isfile(pathArchivo)):
            session = pd.read_csv(pathArchivo, header=None, names=cabeceras)
            estudiante = session[columnas]
            
            for indice, fila in estudiante.iterrows():
                clave = (fila['activity']).split("_")[0].strip()
                idEstudiante = fila['student_Id']
                if (fila[idEstudiante].strip() in recolectaDatos[idEstudiante]):
                    recolectaDatos[idEstudiante][clave].append(fila['idle_time']/1000)
                else:
                    recolectaDatos[idEstudiante][clave] = []
                    recolectaDatos[idEstudiante][clave].append(fila['idle_time']/1000)
                    
                    
```
