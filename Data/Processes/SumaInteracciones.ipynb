## Probando como sacar la diferencia de los segundos en python

```python
from datetime import datetime


dt_obj = datetime.strptime('20.11.2014 12:21:48','%d.%m.%Y %H:%M:%S')

dt_obj1= datetime.strptime('20.11.2014 12:21:49','%d.%m.%Y %H:%M:%S')

res = dt_obj1-dt_obj
value = res.total_seconds()
print(round(value,3))
```
